````python
# –ú–æ–¥—É–ª—å 2.1: –í–≤—ñ–¥ (`input`) —Ç–∞ –≤–∏–≤—ñ–¥ (`print`) —É Python

> **–§–æ—Ä–º–∞—Ç:** –∫–æ–Ω—Å–ø–µ–∫—Ç —É —Å—Ç–∏–ª—ñ Jupyter Notebook ‚Äî —á–µ—Ä–≥—É—é—Ç—å—Å—è Markdown‚Äë–ø–æ—è—Å–Ω–µ–Ω–Ω—è —Ç–∞ –∫–æ–¥‚Äë–æ—Å–µ—Ä–µ–¥–∫–∏ –¥–ª—è –∑–∞–ø—É—Å–∫—É.

---

## üéØ –ú–µ—Ç–∞ —Ç–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –Ω–∞–≤—á–∞–Ω–Ω—è

–ü—ñ—Å–ª—è –ø—Ä–æ—Ö–æ–¥–∂–µ–Ω–Ω—è —Ü—å–æ–≥–æ –º–æ–¥—É–ª—è –≤–∏ –∑–º–æ–∂–µ—Ç–µ:

* –ø–æ—è—Å–Ω–∏—Ç–∏, —è–∫ –ø—Ä–∞—Ü—é—î `print()` —Ç–∞ –Ω–∞–ª–∞—à—Ç–æ–≤—É–≤–∞—Ç–∏ –≤–∏–≤—ñ–¥ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ `sep` —ñ `end`;
* –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ `input()` –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –≤—ñ–¥ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ —Ç–∞ –∫–æ—Ä–µ–∫—Ç–Ω–æ –ø–µ—Ä–µ—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ —Ç–∏–ø–∏ (`int`, `float`, `str`);
* —Ñ–æ—Ä–º–∞—Ç—É–≤–∞—Ç–∏ —Ä—è–¥–∫–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é **f-strings** (—Ç–æ—á–Ω—ñ—Å—Ç—å, —à–∏—Ä–∏–Ω–∞, –≤–∏—Ä—ñ–≤–Ω—é–≤–∞–Ω–Ω—è, —Ä–æ–∑–¥—ñ–ª—å–Ω–∏–∫–∏ —Ç–∏—Å—è—á);
* –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –±–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤–∏–º–∏ —Ä—è–¥–∫–∞–º–∏ —Ç–∞ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–∏–º–∏ —Å–∏–º–≤–æ–ª–∞–º–∏ (`\n`, `\t`, `\\`, –ª–∞–ø–∫–∏);
* –Ω–∞–ø–∏—Å–∞—Ç–∏ –ø—Ä–æ—Å—Ç—ñ —ñ–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ñ –º—ñ–Ω—ñ‚Äë–ø—Ä–æ–≥—Ä–∞–º–∏ (–ø—Ä–∏–≤—ñ—Ç–∞–Ω–Ω—è, –∫–æ–Ω–≤–µ—Ä—Ç–µ—Ä–∏, –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä–∏).

> **–ü–µ—Ä–µ–¥—É–º–æ–≤–∏:** –±–∞–∑–æ–≤–∏–π —Å–∏–Ω—Ç–∞–∫—Å–∏—Å Python (–∑–º—ñ–Ω–Ω—ñ, —Ç–∏–ø–∏ –¥–∞–Ω–∏—Ö), –≤–º—ñ–Ω–Ω—è –∑–∞–ø—É—Å–∫–∞—Ç–∏ –∫–ª—ñ—Ç–∏–Ω–∫–∏ –≤ Jupyter/Colab.

---

## üß≠ –ó–º—ñ—Å—Ç

* [–Ø–∫ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –±–ª–æ–∫–Ω–æ—Ç–æ–º](#—è–∫-–ø—Ä–∞—Ü—é–≤–∞—Ç–∏-–∑-–±–ª–æ–∫–Ω–æ—Ç–æ–º)
* [1. `print()` ‚Äî –≤–∏–≤–µ–¥–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö](#1-print--–≤–∏–≤–µ–¥–µ–Ω–Ω—è-–¥–∞–Ω–∏—Ö)
* [2. `input()` ‚Äî –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö](#2-input--–æ—Ç—Ä–∏–º–∞–Ω–Ω—è-–¥–∞–Ω–∏—Ö)
* [3. –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∏–ø—ñ–≤](#3-–ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è-—Ç–∏–ø—ñ–≤)
* [4. –ü—Ä–∞–∫—Ç–∏–∫–∞: –ø—Ä–æ—Å—Ç–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä](#4-–ø—Ä–∞–∫—Ç–∏–∫–∞-–ø—Ä–æ—Å—Ç–∏–π-–∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä)
* [5. –§–æ—Ä–º–∞—Ç–æ–≤–∞–Ω—ñ —Ä—è–¥–∫–∏ (f‚Äëstrings)](#5-—Ñ–æ—Ä–º–∞—Ç–æ–≤–∞–Ω—ñ-—Ä—è–¥–∫–∏-fstrings)
* [6. –ë–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤—ñ —Ä—è–¥–∫–∏](#6-–±–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤—ñ-—Ä—è–¥–∫–∏)
* [7. –°–ø–µ—Ü—ñ–∞–ª—å–Ω—ñ —Å–∏–º–≤–æ–ª–∏ —Ç–∞ raw‚Äë—Ä—è–¥–∫–∏](#7-—Å–ø–µ—Ü—ñ–∞–ª—å–Ω—ñ-—Å–∏–º–≤–æ–ª–∏-—Ç–∞-raw—Ä—è–¥–∫–∏)
* [8. –ú—ñ–Ω—ñ‚Äë–ø—Ä–æ—î–∫—Ç–∏](#8-–º—ñ–Ω—ñ–ø—Ä–æ—î–∫—Ç–∏)
* [–ü—ñ–¥—Å—É–º–æ–∫ + –®–ø–∞—Ä–≥–∞–ª–∫–∞](#–ø—ñ–¥—Å—É–º–æ–∫--—à–ø–∞—Ä–≥–∞–ª–∫–∞)
* [–ö–æ–Ω—Ç—Ä–æ–ª—å–Ω—ñ –ø–∏—Ç–∞–Ω–Ω—è](#–∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ñ-–ø–∏—Ç–∞–Ω–Ω—è)
* [–î–æ–º–∞—à–Ω—è –ø—Ä–∞–∫—Ç–∏–∫–∞](#–¥–æ–º–∞—à–Ω—è-–ø—Ä–∞–∫—Ç–∏–∫–∞)

---

## –Ø–∫ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –±–ª–æ–∫–Ω–æ—Ç–æ–º

1. –ó–∞–ø—É—Å–∫–∞–π—Ç–µ –∫–ª—ñ—Ç–∏–Ω–∫–∏ –∫–æ–¥—É –ø–æ —á–µ—Ä–∑—ñ –∑–≤–µ—Ä—Ö—É –≤–Ω–∏–∑.
2. –Ø–∫—â–æ —â–æ—Å—å ¬´–∑–∞–≤–∏—Å–ª–æ¬ª, –≤–∏–∫–æ–Ω–∞–π—Ç–µ **Kernel ‚Üí Restart** —ñ –∑–∞–ø—É—Å–∫–∞–π—Ç–µ –∑–Ω–æ–≤—É.
3. –Ø–∫—â–æ —Å–µ—Ä–µ–¥–æ–≤–∏—â–µ –Ω–µ –ø—ñ–¥—Ç—Ä–∏–º—É—î —ñ–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–∏–π –≤–≤—ñ–¥, –≤–∏ –º–æ–∂–µ—Ç–µ:

   * –∑–∞–º—ñ–Ω–∏—Ç–∏ `input()` –Ω–∞ –∑–∞–∑–¥–∞–ª–µ–≥—ñ–¥—å –≤–∏–∑–Ω–∞—á–µ–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è,
   * –∞–±–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—é `try/except EOFError` (–ø—Ä–∏–∫–ª–∞–¥–∏ –Ω–∏–∂—á–µ).

---

## 1. `print()` ‚Äî –≤–∏–≤–µ–¥–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö

### üìò –¢–µ–æ—Ä—ñ—è

`print()` –≤–∏–≤–æ–¥–∏—Ç—å –ø–µ—Ä–µ–¥–∞–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —É —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–π –ø–æ—Ç—ñ–∫ –≤–∏–≤–æ–¥—É. –ö–ª—é—á–æ–≤—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏:

* `sep` ‚Äî —Ä–æ–∑–¥—ñ–ª—å–Ω–∏–∫ –º—ñ–∂ –µ–ª–µ–º–µ–Ω—Ç–∞–º–∏ (–∑–∞ –∑–∞–º–æ–≤—á—É–≤–∞–Ω–Ω—è–º –ø—Ä–æ–±—ñ–ª `' '`)
* `end` ‚Äî ¬´—Ö–≤—ñ—Å—Ç¬ª –Ω–∞–ø—Ä–∏–∫—ñ–Ω—Ü—ñ —Ä—è–¥–∫–∞ (–∑–∞ –∑–∞–º–æ–≤—á—É–≤–∞–Ω–Ω—è–º –Ω–æ–≤–∏–π —Ä—è–¥–æ–∫ `"\n"`)

### ‚ñ∂Ô∏è –ö–æ–¥‚Äë–∫–ª—ñ—Ç–∏–Ω–∫–∞

```python
print("=" * 50)
print("1. –û–°–ù–û–í–ò print()")
print("=" * 50)

# –ü—Ä–æ—Å—Ç–∏–π —Ç–µ–∫—Å—Ç
print("–ü—Ä–∏–≤—ñ—Ç, —Å–≤—ñ—Ç!")
print("Hello, World!")

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —á–∏—Å–µ–ª
print(42)
print(3.14)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è –¥–µ–∫—ñ–ª—å–∫–æ—Ö –∑–Ω–∞—á–µ–Ω—å
print("–ú–æ—î —ñ–º'—è:", "–Ü–≤–∞–Ω", "–ú–µ–Ω—ñ", 25, "—Ä–æ–∫—ñ–≤")

# –ü–∞—Ä–∞–º–µ—Ç—Ä sep (separator) ‚Äî —Ä–æ–∑–¥—ñ–ª—å–Ω–∏–∫ –º—ñ–∂ –µ–ª–µ–º–µ–Ω—Ç–∞–º–∏
print("Python", "Java", "JavaScript", sep=" | ")
print("2024", "10", "23", sep="-")  # 2024-10-23

# –ü–∞—Ä–∞–º–µ—Ç—Ä end ‚Äî —â–æ –≤–∏–≤–æ–¥–∏—Ç–∏ –≤ –∫—ñ–Ω—Ü—ñ (—Ç–∏–ø–æ–≤–æ "\n")
print("–ü–µ—Ä—à–∏–π —Ä—è–¥–æ–∫", end=" ")
print("–¢–æ–π —Å–∞–º–∏–π —Ä—è–¥–æ–∫!")  # –í–∏–≤–µ–¥–µ—Ç—å—Å—è –≤ –æ–¥–∏–Ω —Ä—è–¥–æ–∫

# –ü–æ–≤–µ—Ä–Ω–µ–Ω–Ω—è –¥–æ –∑–≤–∏—á–∞–π–Ω–æ—ó –ø–æ–≤–µ–¥—ñ–Ω–∫–∏
print("–ê —Ü–µ –≤–∂–µ –Ω–æ–≤–∏–π —Ä—è–¥–æ–∫")

# –ö–æ–º–±—ñ–Ω–∞—Ü—ñ—è sep —Ç–∞ end
print("–†–∞—Ö—É–Ω–æ–∫:", 1, 2, 3, sep="-", end=" | ")
print("–ü—Ä–æ–¥–æ–≤–∂–µ–Ω–Ω—è")

print()  # –ü—É—Å—Ç–∏–π —Ä—è–¥–æ–∫ –¥–ª—è –≤—ñ–¥—Å—Ç—É–ø—É
```

### ‚úÖ –ü–æ—Ä–∞–¥–∏

* –î–ª—è –ª–æ–≥—ñ–≤ –∑—Ä—É—á–Ω–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ —è–≤–Ω—ñ —Ä–æ–∑–¥—ñ–ª—å–Ω–∏–∫–∏: `sep=","` –∞–±–æ `sep=" | "`.
* –£–Ω–∏–∫–∞–π—Ç–µ –∫–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü—ñ—ó —á–∏—Å–µ–ª —ñ —Ä—è–¥–∫—ñ–≤ —á–µ—Ä–µ–∑ `+` ‚Äî –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –∫–æ–º–∏ –∞–±–æ f‚Äëstrings.

### üß™ –°–∞–º–æ–ø–µ—Ä–µ–≤—ñ—Ä–∫–∞

–°–ø—Ä–æ–±—É–π—Ç–µ –∑–∞ –æ–¥–∏–Ω –≤–∏–∫–ª–∏–∫ `print()` –≤–∏–≤–µ—Å—Ç–∏ –¥–∞—Ç—É –≤ —Ñ–æ—Ä–º–∞—Ç—ñ `YYYY/MM/DD`, –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–≤—à–∏ `sep`.

---

## 2. `input()` ‚Äî –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö

### üìò –¢–µ–æ—Ä—ñ—è

* `input(prompt)` –ø–æ–∫–∞–∑—É—î –ø—ñ–¥–∫–∞–∑–∫—É —Ç–∞ **–∑–∞–≤–∂–¥–∏ –ø–æ–≤–µ—Ä—Ç–∞—î —Ä—è–¥–æ–∫** (`str`).
* –ß–∞—Å—Ç–æ –ø–æ—Ç—Ä—ñ–±–Ω–æ –ø–µ—Ä–µ—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ —Ä—è–¥–æ–∫ –Ω–∞ —á–∏—Å–ª–æ: `int(...)`, `float(...)`.

### ‚ñ∂Ô∏è –ë–∞–∑–æ–≤–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è

```python
# –†–æ–∑–∫–æ–º–µ–Ω—Ç—É–π—Ç–µ –¥–ª—è —ñ–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–æ—ó –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏
# name = input("–í–≤–µ–¥—ñ—Ç—å –≤–∞—à–µ —ñ–º'—è: ")
# print("–ü—Ä–∏–≤—ñ—Ç,", name)

# age_str = input("–°–∫—ñ–ª—å–∫–∏ –≤–∞–º —Ä–æ–∫—ñ–≤? ")
# print("–í–∏ –≤–≤–µ–ª–∏:", age_str)
# print("–¢–∏–ø –¥–∞–Ω–∏—Ö:", type(age_str))  # <class 'str'>
```

### ‚ö†Ô∏è –¢–∏–ø–æ–≤—ñ –ø–æ–º–∏–ª–∫–∏

* `ValueError` –ø—Ä–∏ `int("19.5")` –∞–±–æ `int("abc")`. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ `float()` –∞–±–æ –ø–µ—Ä–µ–≤—ñ—Ä–∫—É –≤–≤–µ–¥–µ–Ω–Ω—è.
* –ü–æ—Ä–æ–∂–Ω—ñ–π –≤–≤—ñ–¥. –î–æ–¥–∞–≤–∞–π—Ç–µ –≤–∞–ª—ñ–¥–∞—Ü—ñ—é (—Ü–∏–∫–ª `while`, –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ `str.isdigit()` —Ç–æ—â–æ).

---

## 3. –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∏–ø—ñ–≤

### üìò –¢–µ–æ—Ä—ñ—è

* `int("25") ‚Üí 25`
* `float("19.99") ‚Üí 19.99`
* `str(42) ‚Üí "42"`

### ‚ñ∂Ô∏è –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è

```python
print("=" * 50)
print("3. –ü–ï–†–ï–¢–í–û–†–ï–ù–ù–Ø –¢–ò–ü–Ü–í")
print("=" * 50)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥–∫–∞ –≤ —á–∏—Å–ª–æ
age_str = "25"
age_int = int(age_str)
print(f"–†—è–¥–æ–∫: {age_str}, —Ç–∏–ø: {type(age_str)}")
print(f"–ß–∏—Å–ª–æ: {age_int}, —Ç–∏–ø: {type(age_int)}")

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥–∫–∞ —É —á–∏—Å–ª–æ –∑ –ø–ª–∞–≤–∞—é—á–æ—é —Ç–æ—á–∫–æ—é
price_str = "19.99"
price_float = float(price_str)
print(f"–†—è–¥–æ–∫: {price_str}, —Ç–∏–ø: {type(price_str)}")
print(f"Float: {price_float}, —Ç–∏–ø: {type(price_float)}")

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —á–∏—Å–ª–∞ –≤ —Ä—è–¥–æ–∫
number = 42
text = str(number)
print(f"–ß–∏—Å–ª–æ: {number}, —Ç–∏–ø: {type(number)}")
print(f"–†—è–¥–æ–∫: {text}, —Ç–∏–ø: {type(text)}")
```

### üß© –ó–∞–¥–∞—á–∞‚Äë–º—ñ–Ω—ñ

–ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `safe_int(prompt)`, —è–∫–∞ –∑–∞–ø–∏—Ç—É—î —á–∏—Å–ª–æ –¥–æ —Ç–∏—Ö –ø—ñ—Ä, –ø–æ–∫–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –Ω–µ –≤–≤–µ–¥–µ –∫–æ—Ä–µ–∫—Ç–Ω–µ —Ü—ñ–ª–µ.

<details>
<summary>–ú–æ–∂–ª–∏–≤–µ —Ä—ñ—à–µ–Ω–Ω—è</summary>

```python
def safe_int(prompt="–í–≤–µ–¥—ñ—Ç—å —Ü—ñ–ª–µ —á–∏—Å–ª–æ: "):
    while True:
        s = input(prompt)
        if s.strip().lstrip("-+").isdigit():
            return int(s)
        print("–°–ø—Ä–æ–±—É–π—Ç–µ —â–µ —Ä–∞–∑ ‚Äî —Ü–µ –º–∞—î –±—É—Ç–∏ —Ü—ñ–ª–µ —á–∏—Å–ª–æ.")
```

</details>

---

## 4. –ü—Ä–∞–∫—Ç–∏–∫–∞: –ø—Ä–æ—Å—Ç–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä

### üìò –Ü–¥–µ—è

–ü–æ–ø—Ä–æ—Å–∏—Ç–∏ —É –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –¥–≤–∞ —á–∏—Å–ª–∞, —Å–∫–ª–∞—Å—Ç–∏ —Ç–∞ –∫—Ä–∞—Å–∏–≤–æ –≤–∏–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

### ‚ñ∂Ô∏è –í–∞—Ä—ñ–∞–Ω—Ç —ñ–∑ –¥–µ–º–æ‚Äë–∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ (–∑—Ä—É—á–Ω–æ, —è–∫—â–æ `input()` –Ω–µ–¥–æ—Å—Ç—É–ø–Ω–∏–π)

```python
print("=" * 50)
print("4. –ü–†–ê–ö–¢–ò–ö–ê: –ü–†–û–°–¢–ò–ô –ö–ê–õ–¨–ö–£–õ–Ø–¢–û–†")
print("=" * 50)

def simple_calculator():
    """–ü—Ä–∏–∫–ª–∞–¥ —Ä–æ–±–æ—Ç–∏ –∑ input —Ç–∞ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è–º —Ç–∏–ø—ñ–≤"""

    print("\nüßÆ –ö–∞–ª—å–∫—É–ª—è—Ç–æ—Ä –¥–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è –¥–≤–æ—Ö —á–∏—Å–µ–ª\n")

    # –Ø–∫—â–æ –ø–æ—Ç—Ä—ñ–±–Ω–∞ –ø–æ–≤–Ω–∞ —ñ–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ñ—Å—Ç—å ‚Äî —Ä–æ–∑–∫–æ–º–µ–Ω—Ç—É–π—Ç–µ:
    # num1 = float(input("–í–≤–µ–¥—ñ—Ç—å –ø–µ—Ä—à–µ —á–∏—Å–ª–æ: "))
    # num2 = float(input("–í–≤–µ–¥—ñ—Ç—å –¥—Ä—É–≥–µ —á–∏—Å–ª–æ: "))

    # –î–µ–º–æ-–∑–Ω–∞—á–µ–Ω–Ω—è (–ø—Ä–∞—Ü—é—î —Ç–∞–º, –¥–µ input –∑–∞–±–æ—Ä–æ–Ω–µ–Ω–∏–π)
    num1 = 10.5
    num2 = 5.3

    result = num1 + num2
    print(f"\n{num1} + {num2} = {result}")
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç (2 –∑–Ω–∞–∫–∏ –ø—ñ—Å–ª—è –∫–æ–º–∏): {result:.2f}")

# –†–æ–∑–∫–æ–º–µ–Ω—Ç—É–π—Ç–µ –¥–ª—è –∑–∞–ø—É—Å–∫—É –≤ —ñ–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–æ–º—É —Å–µ—Ä–µ–¥–æ–≤–∏—â—ñ:
# simple_calculator()
```

### üîß –í–∞—Ä—ñ–∞–Ω—Ç –∑ –æ–±—Ä–æ–±–∫–æ—é –ø–æ–º–∏–ª–æ–∫ –≤–≤–æ–¥—É

```python
def calculator_with_validation():
    def ask_float(msg):
        while True:
            try:
                return float(input(msg))
            except ValueError:
                print("–í–≤–µ–¥—ñ—Ç—å —á–∏—Å–ª–æ (–º–æ–∂–Ω–∞ –∑ –∫—Ä–∞–ø–∫–æ—é).")

    a = ask_float("–ü–µ—Ä—à–µ —á–∏—Å–ª–æ: ")
    b = ask_float("–î—Ä—É–≥–µ —á–∏—Å–ª–æ: ")
    print(f"–°—É–º–∞: {a + b:.2f}")
```

---

## 5. –§–æ—Ä–º–∞—Ç–æ–≤–∞–Ω—ñ —Ä—è–¥–∫–∏ (f‚Äëstrings)

### üìò –¢–µ–æ—Ä—ñ—è

* –Ü–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü—ñ—è: `f"–ú–µ–Ω–µ –∑–≤–∞—Ç–∏ {name}, –º–µ–Ω—ñ {age} —Ä–æ–∫—ñ–≤"`
* –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª: `:.2f` (—Ç–æ—á–Ω—ñ—Å—Ç—å), `:10.2f` (—à–∏—Ä–∏–Ω–∞ + —Ç–æ—á–Ω—ñ—Å—Ç—å), `:,` (—Ä–æ–∑–¥—ñ–ª—å–Ω–∏–∫ —Ç–∏—Å—è—á), `:.2%` (–≤—ñ–¥—Å–æ—Ç–∫–∏)

### ‚ñ∂Ô∏è –ü—Ä–∏–∫–ª–∞–¥–∏

```python
print("=" * 50)
print("5. F-STRINGS (—Ñ–æ—Ä–º–∞—Ç–æ–≤–∞–Ω—ñ —Ä—è–¥–∫–∏)")
print("=" * 50)

name = "–û–ª–µ–∫—Å—ñ–π"
age = 28
height = 1.82

# –°—Ç–∞—Ä—ñ –ø—ñ–¥—Ö–æ–¥–∏
print("–ú–µ–Ω–µ –∑–≤–∞—Ç–∏ " + name + ", –º–µ–Ω—ñ " + str(age) + " —Ä–æ–∫—ñ–≤")
print("–ú–µ–Ω–µ –∑–≤–∞—Ç–∏ {}, –º–µ–Ω—ñ {} —Ä–æ–∫—ñ–≤".format(name, age))

# –°—É—á–∞—Å–Ω–∏–π —Å–ø–æ—Å—ñ–± ‚Äî f-strings
print(f"–ú–µ–Ω–µ –∑–≤–∞—Ç–∏ {name}, –º–µ–Ω—ñ {age} —Ä–æ–∫—ñ–≤")
print(f"–ß–µ—Ä–µ–∑ 5 —Ä–æ–∫—ñ–≤ –º–µ–Ω—ñ –±—É–¥–µ {age + 5} —Ä–æ–∫—ñ–≤")
print(f"–ú—ñ–π –∑—Ä—ñ—Å—Ç: {height} –º –∞–±–æ {height * 100} —Å–º")

price = 1234.5
print(f"–¶—ñ–Ω–∞: {price:.2f} –≥—Ä–Ω")       # 2 –∑–Ω–∞–∫–∏ –ø—ñ—Å–ª—è –∫–æ–º–∏
print(f"–¶—ñ–Ω–∞: {price:10.2f} –≥—Ä–Ω")     # —à–∏—Ä–∏–Ω–∞ 10 —Å–∏–º–≤–æ–ª—ñ–≤
print(f"–¶—ñ–Ω–∞: {price:,.2f} –≥—Ä–Ω")      # —Ä–æ–∑–¥—ñ–ª—å–Ω–∏–∫ —Ç–∏—Å—è—á

pi = 3.14159265359
print(f"–ß–∏—Å–ª–æ –ø—ñ: {pi:.2f}")
print(f"–ß–∏—Å–ª–æ –ø—ñ: {pi:.4f}")

ratio = 0.0785
print(f"–ß–∞—Å—Ç–∫–∞: {ratio:.2%}")  # 7.85%

# –í–∏—Ä—ñ–≤–Ω—é–≤–∞–Ω–Ω—è
print(f"|{123:>6}| –ø—Ä–∞–≤–æ—Ä—É—á")
print(f"|{123:<6}| –ª—ñ–≤–æ—Ä—É—á")
print(f"|{123:^6}| –ø–æ —Ü–µ–Ω—Ç—Ä—É")
```

### üí° –ü–æ—Ä–∞–¥–∞

–î–ª—è —Å–∫–ª–∞–¥–Ω–∏—Ö —Ñ–æ—Ä–º–∞—Ç—ñ–≤ –¥–∏–≤—ñ—Ç—å—Å—è ¬´Format Specification Mini‚ÄëLanguage¬ª (Python docs).

---

## 6. –ë–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤—ñ —Ä—è–¥–∫–∏

### üìò –¢–µ–æ—Ä—ñ—è

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ø–æ—Ç—Ä—ñ–π–Ω—ñ –ª–∞–ø–∫–∏ `"""..."""` –∞–±–æ `'''...'''` –¥–ª—è –±–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤–∏—Ö –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω—å/—à–∞–±–ª–æ–Ω—ñ–≤. –á—Ö –∑—Ä—É—á–Ω–æ –∫–æ–º–±—ñ–Ω—É–≤–∞—Ç–∏ –∑ f‚Äëstrings.

### ‚ñ∂Ô∏è –ü—Ä–∏–∫–ª–∞–¥–∏

```python
print("=" * 50)
print("6. –ë–ê–ì–ê–¢–û–†–Ø–î–ö–û–í–Ü –†–Ø–î–ö–ò")
print("=" * 50)

message = """
–¶–µ –±–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤–∏–π —Ç–µ–∫—Å—Ç.
–í—ñ–Ω –º–æ–∂–µ –º—ñ—Å—Ç–∏—Ç–∏ –¥–µ–∫—ñ–ª—å–∫–∞ —Ä—è–¥–∫—ñ–≤.
–ó—Ä—É—á–Ω–æ –¥–ª—è –¥–æ–≤–≥–∏—Ö –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω—å!
"""
print(message)

name = "–ú–∞—Ä—ñ—è"
greeting = f"""
–ü—Ä–∏–≤—ñ—Ç, {name}!

–¶–µ –±–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤–µ –ø—Ä–∏–≤—ñ—Ç–∞–Ω–Ω—è.
–°–ø–æ–¥—ñ–≤–∞—î–º–æ—Å—å, —É —Ç–µ–±–µ —á—É–¥–æ–≤–∏–π –¥–µ–Ω—å!
"""
print(greeting)
```

> **Note:** –Ø–∫—â–æ –≤–∏ –≤—Å—Ç–∞–≤–ª—è—î—Ç–µ –±–∞–≥–∞—Ç–æ—Ä—è–¥–∫–æ–≤–∏–π —Ç–µ–∫—Å—Ç –≤—Å–µ—Ä–µ–¥–∏–Ω—É –∫–æ–¥—É –∑ –≤—ñ–¥—Å—Ç—É–ø–∞–º–∏, –∑–∞–π–≤—ñ –ø—Ä–æ–±—ñ–ª–∏ –º–æ–∂–Ω–∞ –ø—Ä–∏–±—Ä–∞—Ç–∏ —á–µ—Ä–µ–∑ `textwrap.dedent`.

---
````

````python

```python
def greet_user():
    """–ü—Ä–∏–≤—ñ—Ç–∞—Ç–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –∑–∞ —ñ–º'—è–º"""
    # name = input("–Ø–∫ –≤–∞—Å –∑–≤–∞—Ç–∏? ")
    # print(f"–ü—Ä–∏—î–º–Ω–æ –ø–æ–∑–Ω–∞–π–æ–º–∏—Ç–∏—Å—å, {name}!")

    # –î–ª—è –¥–µ–º–æ –±–µ–∑ input:
    name = "–ê–Ω–¥—Ä—ñ–π"
    print(f"–ü—Ä–∏—î–º–Ω–æ –ø–æ–∑–Ω–∞–π–æ–º–∏—Ç–∏—Å—å, {name}!")

print("\n--- –ü—Ä–∏–∫–ª–∞–¥ 1: –ü—Ä–∏–≤—ñ—Ç–∞–Ω–Ω—è ---")
greet_user()
```

### 8.2 –í—ñ–∫ —É –¥–Ω—è—Ö

```python
def calculate_age_in_days():
    """–†–æ–∑—Ä–∞—Ö—É–≤–∞—Ç–∏ –≤—ñ–∫ —É –¥–Ω—è—Ö (‚âà –±–µ–∑ —É—Ä–∞—Ö—É–≤–∞–Ω–Ω—è –≤–∏—Å–æ–∫–æ—Å–Ω–∏—Ö —Ä–æ–∫—ñ–≤)"""
    # age = int(input("–°–∫—ñ–ª—å–∫–∏ –≤–∞–º —Ä–æ–∫—ñ–≤? "))

    # –î–ª—è –¥–µ–º–æ:
    age = 25

    days = age * 365
    print(f"–í–∞–º –ø—Ä–∏–±–ª–∏–∑–Ω–æ {days} –¥–Ω—ñ–≤!")

print("\n--- –ü—Ä–∏–∫–ª–∞–¥ 2: –í—ñ–∫ —É –¥–Ω—è—Ö ---")
calculate_age_in_days()
```
````

````python

```python
def temperature_converter():
    """–ö–æ–Ω–≤–µ—Ä—Ç–µ—Ä —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∏ –¶–µ–ª—å—Å—ñ–π ‚Üí –§–∞—Ä–µ–Ω–≥–µ–π—Ç"""
    # celsius = float(input("–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –≤ –¶–µ–ª—å—Å—ñ—è—Ö: "))

    # –î–ª—è –¥–µ–º–æ:
    celsius = 25

    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}¬∞C = {fahrenheit}¬∞F")

print("\n--- –ü—Ä–∏–∫–ª–∞–¥ 3: –ö–æ–Ω–≤–µ—Ä—Ç–µ—Ä —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∏ ---")
temperature_converter()
```

---
````
